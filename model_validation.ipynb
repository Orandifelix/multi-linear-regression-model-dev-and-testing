### MODEL EVALUATION  
- TESTING OVERFITING 
- UNDERFITTING 


```python
import pandas  as pd 
from  sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
data = pd.read_csv(url)
data.head(3)
```

- checking for  missing values 

```python
data.isnull().sum()
```

```python
#Filling Null values with  mean
data['horsepower'] = data['horsepower'].fillna(data['horsepower'].mean())
```

- Selecting feature and target for out  model 

```python
X = data[['horsepower']]
y= data['mpg']
```

```python
# Splitting the data for  train and  testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
print(X_test.shape, y_test.shape)
print(X_train.shape, y_train.shape)
```

```python
#Model  fitting 
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# Predicting  the values 
y_pred = model.predict(X_test)

print(y_pred[:5])  # print first 5 predictions
```

```python
print(y_test[:5])  # print first 5  -actual values 
```

```python
#checking the metrics 

mse = mean_squared_error(y_test, y_pred) # comparing the actual and predicted values

print("Mean Squared Error:", mse)
```

```python
y_train_pred = model.predict(X_train) # prediction should match the y_train values
```

```python
#checking overfitting and underfitting

# Calculating true RMSE
rmse_training = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Training RMSE: {rmse_training}")
print(f"Test RMSE: {rmse_test}")
```

```python
from sklearn.model_selection import cross_validate, cross_val_score

csv_score = cross_val_score(model, X, y,  cv =5, scoring= 'neg_mean_squared_error')

rmse_scores = np.sqrt(-csv_score)
print("RMSE Scores:", rmse_scores)
print("Average RMSE:", rmse_scores.mean())
```

```python
cvs_score =cross_val_score(model,X,y,cv=5,scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-cvs_score)
print("RMSE Scores:", rmse_scores)
print("Average RMSE:", rmse_scores.mean())
```

### Regulazation

```python
# Feature and  target

y= data['mpg']

X = data.drop(['mpg', 'origin', 'name'], axis=1)
```

```python
#Splitting 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=10)
```

```python
# Standardization  the features 
scalar = StandardScaler()
X_train_s  = scalar.fit_transform(X_train)
X_test_s = scalar.fit_transform(X_test)
```

```python
 # Building  the  models 
 
lin = LinearRegression()
lin.fit(X_train_s, y_train)

#ridge regression
ridge = Ridge(alpha = 0.5)
ridge.fit(X_train_s, y_train)

#Lasso regression
Lasso = Lasso(alpha = 0.5)
Lasso.fit(X_train_s, y_train)
```

```python
#prediction
lin_train = lin.predict(X_train_s)
lin_test = lin.predict(X_test_s)

ridge_train = ridge.predict(X_train_s)
ridge_test = ridge.predict(X_test_s)

Lasso_train = Lasso.predict(X_train_s)
Lasso_test = Lasso.predict(X_test_s)
```

```python
print("train error lin:", mean_squared_error(lin_train,y_train))
print("test error lin:", mean_squared_error(lin_test,y_test))
print("\n")
print("train error ridge:", mean_squared_error(ridge_train,y_train))
print("test error ridge:", mean_squared_error(ridge_test,y_test))
print("\n")
print("train error Lasso:", mean_squared_error(Lasso_train,y_train))
print("test error Lasso:", mean_squared_error(Lasso_test,y_test))
```

```python
print("lin parameter coef:", lin.coef_)
print("ridge parameter coef:", ridge.coef_)
print("Lasso parameter coef:", Lasso.coef_)
```
