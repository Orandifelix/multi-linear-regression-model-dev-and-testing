```python
class BankAccount:
    def __init__(self, acc_no, balance): # instance method 
        self.acc_no = acc_no  # attributes
        self.balance = balance # attributes 
    def deposit(self, amount):
        self.balance += amount
        
    def withdraw(self, amount):
        if amount > self.balance:
            raise  Exception("Insuffiecient balance")
        else:
            self.balance -= amount
            return self.balance
        
    def check_balance(self):
        return self.balance
```

```python
my_bank_acc = BankAccount("01100000001", 1000)
```

```python
print(my_bank_acc.acc_no)
print(my_bank_acc.balance)
```

```python
my_bank_acc.deposit(1000)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import pandas as pd
import  numpy as np
```

```python
class reg_model:
    def __init__(self, X, y, model):
        self.X = X
        self.y = y
        self.model = model
        self.scaler = StandardScaler() # tool 
    
    # splitting data
    def split_data(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y,  test_size=0.2, random_state=7)
        
    # preprocessing 
    def preprocess(self):
        self.X_train_scaled = self.scaler.fit_transform(self.X_train)
    
    # training 
    def train(self):
        self.model.fit(self.X_train_scaled, self.y_train)
        
    #predicting 
    def predict(self):
        self.y_pred = self.model.predict(self.X_test)
        return self.y_pred
```

```python
# Sample data  
df = pd.DataFrame({
    "feature1": [1,2,3,4,5],
    "feature2": [10,20,30,40,50],
    "target": [3,6,9,12,15]
})

X = df.drop(columns=['target'])
y = df['target']
```

```python
#regression
model = LinearRegression()

reg = reg_model(X, y, model)
reg.split_data()
reg.preprocess()
reg.train()
reg.predict()
```

```python
#classification 
model = LogisticRegression()

reg = reg_model(X, y, model)
reg.split_data()
reg.preprocess()
reg.train()
reg.predict()
```
